```
import yfinance as yf

msft = yf.Ticker("XLF")

# get stock info
print(msft.info)

# get historical market data
hist = msft.history(period="30y", interval="1d")

#print(type(hist))
#print(hist)
#print(hist[hist['Dividends']>0])
#print(hist.index("1993-03-19"))

print(hist.index[0])
print(hist.index[-1])
#print(type(hist.index[-1]))

delta = hist.index[-1] - hist.index[0]
print(delta.days / 30)

step = 2
InvestedMoney=0
endMoney = 0
shares = 0
moneyPerMonth =1000
moneyPerDay = moneyPerMonth/30*delta.days/len(hist.index)
print(delta.days/len(hist.index))
print(moneyPerDay)
#print((hist.iloc[0][1]))

for i in range(0, len(hist.index), step):
        InvestedMoney+=moneyPerDay*step
        print(hist.iloc[i][0])
        shares += (moneyPerDay*step)/(hist.iloc[i][0])

endMoney = shares*hist.iloc[-1][0]
    # print("END RESULTS:")
print("Invested money ", InvestedMoney)
print("End Money ", endMoney)
interestRate = (endMoney-InvestedMoney)/InvestedMoney*100
print("Rate of interest: ", interestRate , " Step days: ", step) 

#hist.to_csv(r'#File Name.csv')
```

```
#! python3
# removeCsvHeader.py - Removes the header from all CSV files in the current
# working directory.

import csv, os , time
from datetime import datetime

def main():
    for csvFilename in os.listdir('.'):
        if not csvFilename.endswith('.csv'):
            continue # skip non-csv files
        print()
        print('Working on file ' + csvFilename + '...')
        getInterestRate(csvFilename,1000,"1980-01-01",3)
        getInterestRate(csvFilename,1000,"1980-01-01",7)
        getInterestRate(csvFilename,1000,"1980-01-01",15)
        getInterestRate(csvFilename,1000,"1980-01-01",30)
        getInterestRate(csvFilename,1000,"1980-01-01",60)
        getInterestRate(csvFilename,1000,"1980-01-01",90) 
        getInterestRate(csvFilename,1000,"1980-01-01",180)
        getInterestRate(csvFilename,1000,"1980-01-01",360)
        getInterestRate(csvFilename,1000,"1980-01-01",720)
        getInterestRate(csvFilename,1000,"1980-01-01",360*3)
        getInterestRate(csvFilename,1000,"1980-01-01",360*4)
        getInterestRate(csvFilename,1000,"1980-01-01",360*5)
        getInterestRate(csvFilename,1000,"1980-01-01",360*6)
        print("********************************************************************")


def drange(start, stop, step):
    while start < stop:
            yield start
            start += step


#startdate in format 2015-12-01 (year-month-day)
def getInterestRate(csvFilename, moneyPerMonth,startdate,step=30):
    csvFileObj = open(csvFilename)
    readerObj = csv.reader(csvFileObj)
 
    data = [row for row in readerObj]
    x = len(data[0])
    y = len(data)

    startdate = data[1][0]
    enddate = data[y-1][0]

    convStart=datetime.strptime(startdate,"%Y-%m-%d")
    # print(convStart)

    # print("first cell ", data[1][0])
    # print("width : ", x)
    # print("height  : ", y)
    # print("startdate : ", startdate)
    # print("enddate : ", enddate)
    
    convEnd=datetime.strptime(enddate,"%Y-%m-%d")
    # print(convEnd)

    daysDelta = (convEnd-convStart)
    # print ("daysDelta :", daysDelta.days)

    InvestedMoney=0
    endMoney = 0
    shares = 0

    
    steprow =int(round(step*(y/daysDelta.days),0))
    # print("step : ", steprow)
    moneyPerDay = moneyPerMonth/30*daysDelta.days/y
    # print("moneyPerDay :", moneyPerDay)
    for i in drange(1, y, steprow):
        InvestedMoney+=moneyPerDay*steprow
        shares += (moneyPerDay*steprow)/float(data[i][4])
        #print(data[i][0]," newmoney: ", moneyPerDay*steprow, " InvestedMoney :", InvestedMoney, "shares :", shares)
        
    endMoney = shares*float(data[y-1][4])
    # print("END RESULTS:")
    print("Invested money ", InvestedMoney)
    print("End Money ", endMoney)
    interestRate = (endMoney-InvestedMoney)/InvestedMoney*100
    print("Rate of interest: ", interestRate , " Step days: ", step) 
    print()
    return interestRate



if __name__ == '__main__': main()
```
